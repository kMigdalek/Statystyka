```python
# zad 1.1


import random


# Girl = 0
# Boy = 1

random.seed(100)

def siblings_calculator(example):
    siblings = []

    # example a)
    if example == 0:
        siblings.append(0)
        siblings.append(random.randint(0, 1))
        if (siblings[0] == 0 and siblings[1] == 0):
            return True
        else: return False


    # example b)
    if example == 1:
        siblings.append(random.randint(0, 1))
        siblings.append(random.randint(0, 1))

        if siblings[0] == 1 and siblings[1] == 1:
            return 
        elif siblings[0] == 0 and siblings[1] == 0:
            return True
        else: return False

    
     # example c)
    if example == 2:
        siblings.append(0)
        siblings.append(random.randint(0, 1))
        if (siblings[0] == 0 and siblings[1] == 0):
            return True
        else: return False
    
    
def prediction_model(example, test_number):
    win = 0
    lose = 0
    for i in range(test_number):
        result = siblings_calculator(example) 
        if result == True:
            win += 1
        elif result == False: 
            lose += 1
    end_result = win/(win + lose) * 100
    return (f'{round((end_result),2)}%')


print(f'a) {prediction_model(0, 10000000)}')
print(f'b) {prediction_model(1, 10000000)}')
print(f'c) {prediction_model(2, 10000000)}')
```

```python
# zad 1.2

# Change variables at the end of the code as needed
# simulation({1-3}, {'yes/not'})

import random

def game(first_doors, change='no'):
   
    values = [True, False, False]
    keys = [1, 2, 3]

    random.shuffle(values)
    doors =dict(zip(keys, values))


    for key, value in doors.items():
        if key != first_doors and value == False:
            shown_doors = key
            break


    for key, value in doors.items():
        if key != first_doors and key !=shown_doors:
            prompted_doors = key
            break



    if change.lower() == 'yes':
        change = prompted_doors
    else: change = first_doors

    if (int(change) == prompted_doors or int(change) == first_doors):
        if doors.get(int(change)) == True:
            return True
        else: return False
           

def simulation(doors, change='no'):
    win = 0
    lose = 0
    if str(doors).isdigit() and 1 <= int(doors) <= 3:
        for i in range(100000):
            if game(int(doors), change) == True:
                win += 1
            else: lose +=1
        print(f'{round(((win/(win+lose))*100),2)}%')
    else: print('Wrong door number')


simulation(1, 'YES')
```

```python
# zad 1.8

# orzeł - 1
# reszka - 0


values = [[0, 1],[0, 1] ,[1,1] ]
keys = [1, 2, 3]

coins = dict(zip(keys, values))

# a)

heads = 0
tails = 0


for i in range(100000):
    random_coin = coins.get(random.randint(1, 3))
    if random_coin[random.randint(0, 1)] == 1:
        heads += 1
    else: tails += 1

result = (heads/(heads + tails)) * 100

print(f'a) {round((result),2)}%')


# b)


heads_false = 0
tails = 0

for i in range(100000):
    random_coin = coins.get(random.randint(1, 3))
    if random_coin == [1, 1]:
        heads_false += 1
    else:
        if random_coin[random.randint(0, 1)] == 1:
            continue
        else: tails += 1

result = (heads_false/(tails + heads_false)) * 100

print(f'b) {round((result),2)}%')
```

```python
# zad 8.4

from scipy import stats
import math


def arithmetic_mean_distribution(u, o, n):

    ox = o/math.sqrt(n)
    print(f'a) N ({u},{int(ox)})')

    # ox will be used to calculate probalibity in the next function
    return ox



def weight_probalibity(u, x):

    z = (x - u)/arithmetic_mean_distribution(u, o, n)
    p = round(((1 - stats.norm.cdf(z)) * 100), 2)

    print(f'b) P(X > 52) = {p}%')


u = 50
o = 3
n = 9

x = 52

weight_probalibity(u, x)
```

```python
# zad 8.5

from scipy import stats
import math

def confidence_interval(x, s, n, a):
    se = s/math.sqrt(n)
    
    # critical value calculated for t-student
    t = stats.t.ppf(1-((a/100)/2), n-1)
    

    lb = round((x - t * se), 2)
    ub = round((x + t * se), 2)

    print(f'The confidence interval is ({lb}%, {ub}%)')


x = 10.37
s = 3.5
n = 15
a = 5

confidence_interval(x, s, n, a)
```

```python
# zad 8.8

from scipy import stats
import math


def required_population(o, e, a):
    z = stats.norm.ppf(1-((a/100)/2))
    n = ((z*o)/e)**2
    print(f'Minimum required population is {round(n)}')


o = 400
e = 120
a = 5

required_population(o, e, a)
```

```python
# zad 9.2

from scipy import stats
import math

def coin_flip(n, x, h, a, p = False):
    u = 100 * h
    o = math.sqrt(n * h * (1 - h))
    z = (x -u)/o

    za = stats.norm.ppf(1 - a)

    # p == True turn on c) example
    if p == True:
        p = round((1 - stats.norm.cdf(z)), 4)
        return p

    
    if z > za:
        return f'{z} > {round((za),2)} Null hypothesis rejected'
    else: return f'{z} < {round((za),2)} Null hypothesis accepted'
    
    
    

n = 100
x = 60
h = 0.5
a = 0.05


print(f'a) {coin_flip(n, x, h, a)}')
print(f'b) {coin_flip(n, x, h, 0.01)}')
print(f'c) {coin_flip(n, x, h, a, True)}')
```
